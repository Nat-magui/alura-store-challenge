#1. Análisis de facturación

````python
facturacion_tienda1 = tienda["Precio"].sum()
facturacion_tienda2 = tienda2["Precio"].sum()
facturacion_tienda3 = tienda3["Precio"].sum()
facturacion_tienda4 = tienda4["Precio"].sum()

facturaciones = {
    "Tienda 1": facturacion_tienda1,
    "Tienda 2": facturacion_tienda2,
    "Tienda 3": facturacion_tienda3,
    "Tienda 4": facturacion_tienda4
}

facturaciones
````

````python
import pandas as pd

facturacion_tienda1 = tienda["Precio"].sum()
facturacion_tienda2 = tienda2["Precio"].sum()
facturacion_tienda3 = tienda3["Precio"].sum()
facturacion_tienda4 = tienda4["Precio"].sum()

facturaciones = {
    "Tienda 1": float(facturacion_tienda1),
    "Tienda 2": float(facturacion_tienda2),
    "Tienda 3": float(facturacion_tienda3),
    "Tienda 4": float(facturacion_tienda4)
}

df_fact = (
    pd.DataFrame(facturaciones.items(), columns=["Tienda", "Ingreso total"])
      .sort_values("Ingreso total", ascending=False)
      .reset_index(drop=True)
)

df_fact["Ingreso total"] = df_fact["Ingreso total"].map(lambda x: f"${x:,.0f}".replace(",", "."))

df_fact
````

# 3. Calificación promedio de la tienda


##📌 1️⃣ Calcularemos la valoración media por tienda

Como la columna se llama "Calificación", simplemente usamos `.mean()`.

````python
# Calificación promedio por tienda

media_t1 = tienda["Calificación"].mean()
media_t2 = tienda2["Calificación"].mean()
media_t3 = tienda3["Calificación"].mean()
media_t4 = tienda4["Calificación"].mean()

# Guardamos en diccionario
valoraciones = {
    "Tienda 1": media_t1,
    "Tienda 2": media_t2,
    "Tienda 3": media_t3,
    "Tienda 4": media_t4
}

# Mostrar resultados de forma más clara
print("⭐ Valoración promedio por tienda:\n")

for tienda, promedio in valoraciones.items():
    print(f"{tienda}: {promedio:.2f}")
````

> 🔎 El :.2f es solo para mostrar 2 decimales y que quede más prolijo.

##📊 2️⃣ Gráfico de barras

````python
import matplotlib.pyplot as plt

plt.bar(valoraciones.keys(), valoraciones.values())

plt.title("Valoración promedio por tienda")
plt.ylabel("Calificación promedio")
plt.ylim(0, 5)  # porque las calificaciones van de 0 a 5
plt.show()
````

##📊 Conclusión

Al analizar la calificación promedio de los clientes, se observa que:

  * Todas las tiendas tienen una valoración cercana a 4 estrellas, lo que indica un buen nivel general de satisfacción.

  * Tienda 3 presenta la mejor valoración promedio (4.05), seguida muy de cerca por Tienda 2 (4.04).

  * Tienda 4 se mantiene en 4.00.

  * Tienda 1 tiene la valoración más baja (3.98), aunque la diferencia es mínima.

> 📌 Esto indica que no existe un problema grave de satisfacción en ninguna tienda. La experiencia del cliente es relativamente homogénea entre todas.

# 5. Envío promedio por tienda

##🧠 ¿Qué queremos lograr con este paso?

Queremos saber cuánto paga en promedio un cliente por el envío en cada tienda.

Como el dataset tiene muchas compras (filas), y cada compra tiene un valor de envío, lo mejor es calcular un promedio para resumir toda esa información en un solo número por tienda.

````python
# Costo de envío promedio por tienda

envio_t1 = tienda["Costo de envío"].mean()
envio_t2 = tienda2["Costo de envío"].mean()
envio_t3 = tienda3["Costo de envío"].mean()
envio_t4 = tienda4["Costo de envío"].mean()

print("📦 Costo de envío promedio por tienda:\n")

print(f"Tienda 1: {envio_t1:.2f}")
print(f"Tienda 2: {envio_t2:.2f}")
print(f"Tienda 3: {envio_t3:.2f}")
print(f"Tienda 4: {envio_t4:.2f}")
````

###✅ ¿Por qué usamos .mean()?

`.mean()` significa media (promedio).

En Pandas, cuando hacemos:


```Python
tienda["Costo de envío"].mean()
```

estamos pidiendo:

    ✅ “Tomá todos los valores de la columna Costo de envío y calculá el promedio”.

Eso nos ayuda a comparar tiendas rápidamente:

  * si una tienda tiene envío promedio más alto, sus clientes están pagando más por envío.

  * si es más bajo, el envío suele ser más barato.

> Importante: en este desafío, el envío lo paga el cliente, por eso esta métrica puede influir en la decisión de compra (y en el rendimiento de la tienda).

> `f"..."` es un f-string para insertar variables dentro del texto.

> `:.2f` sirve para mostrar solo 2 decimales (más prolijo).

##📊 Gráfico de barras del promedio de envio de cada tienda

````python
import matplotlib.pyplot as plt

envios = [envio_t1, envio_t2, envio_t3, envio_t4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.bar(tiendas, envios)

plt.title("Costo de envío promedio por tienda")
plt.ylabel("Costo promedio de envío")

plt.show()
````

##Conclusion

> El análisis del costo de envío promedio muestra cuánto pagan los clientes, en promedio, por el envío en cada tienda. Comparar estos valores permite identificar si alguna tienda tiene costos de envío significativamente más altos, lo que podría afectar la decisión de compra de los clientes. Esta métrica también ayuda a comprender posibles diferencias logísticas entre las tiendas.
